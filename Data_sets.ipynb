```python
data_na=data.dropna()
```

```python
data_na.info()
```

```python
#convert categorical to the numerical characteristic (proximity to the ocean)
data_na["ocean_proximity"]
```

```python
data_na["ocean_proximity"].value_counts()
#1,2,3,4,5
```

```python
dummy_columns = ['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']
data_na = data_na.drop(columns=dummy_columns, errors='ignore')
```

```python
#Dummies/One-Hot encoding
# NEAR BAY  INLAND  NEAR OCEAN
#    1        0         0
#    0        0         1

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)
data_na = data_na.join(dummies)
```

```python
data_na.head()
```

```python
data_na=data_na.drop(["ocean_proximity"],axis=1)
```

```python
data_na.head()
```

```python
#analysis and addition of new features
```

```python
#corelations
data_na.corr()
```

```python
#used to change the size of the graph
#rc=will be a dictionary ,in which the key is figure.figsize and the value
sb.set(rc={'figure.figsize': (15,8)})
sb.heatmap(data_na.corr(), annot=True, cmap="YlGnBu")
```
