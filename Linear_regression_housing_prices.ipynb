```python
dummy_columns = ['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']
data_na = data_na.drop(columns=dummy_columns, errors='ignore')
```

```python
#Dummies/One-Hot encoding
# NEAR BAY  INLAND  NEAR OCEAN
#    1        0         0
#    0        0         1

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)
data_na = data_na.join(dummies)
```

```python
#add new features depending on the data
#propotion of how many rooms are sleeping rooms
#bedroom ratio for propotion of sleeping rooms/bedrooms
data_na["bedroom_ratio"]=data_na["total_bedrooms"]/data_na["total_rooms"]
```

```python
#fewer of the rooms for are sleeping, the house costs more(maybe becuz it has entertainment rooms, cleaning rooms etc)
```

```python
#seperate characteristics of the label and features(inputs x)
x=data_na.drop(["median_house_value"],axis=1)
y=data_na["median_house_value"]
```

```python
#seperate data in 2 parts 1 for training and 1 for testing - training adjusts and improves
# test with data not seen before
from sklearn.model_selection import train_test_split

#this returns 4 data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
#keys for the dictonary - renaming and adding another column
comparison={"Prediction":predictions,"Actual Value":y_test}
pd.DataFrame(comparison)

#df=pd.DataFrame(predictions)
#using this table show seperate predicted and actual value
#df.head()
```

```python
#model works very well with training data and bad with testing data - overfitting (cant generalise the data)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
# values are close so not overfitting
```

```python
#error- take all data from predictions and actual values and calculate somehow pooled to see how badly the predictions are missed
from sklearn.metrics import mean_squared_error
import numpy as np

mse=mean_squared_error(y_test,predictions)
```

```python
#big value as we are squaring it, the residuals/error
mse
```

```python
rmse=np.sqrt(mse)
```
